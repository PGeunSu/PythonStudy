```python
import pandas as pd

df = pd.read_excel("score.xlsx" ,index_col="지원번호")
df
```

```python
df["학교"] = df["학교"] + "등학교"
df
```

```python
#데이터 함수 적용(apply)
#키에 cm을 붙이는 역할

def add_cm(height):
    return str(height) + "cm"

df["키"] = df["키"].apply(add_cm) #키 데이터에 대해 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): #Nan이 아닌지
        return lang.capitalize() #첫글자는 대문자, 나머지는 소문자
    return lang
df["SW특기"] = df["SW특기"].apply(capitalize)
df
```

```python
#그룹화
df.groupby("학교").get_group("북산고등학교")
```

```python
df.groupby("학교").mean() #계산 가능한 데이터들의 평균값
```

```python
df.groupby("학교").size() #각 그룹의 크기
```

```python
df.groupby("학교").size()["북산고등학교"]
```

```python
df.groupby("학교")["키"].mean() #학교로 그룹화를 한 뒤에 키의 평균 데이터
```

```python
df.groupby("학교")[["국어","영어","수학"]].mean() #학교로 그룹화를 한 뒤에 국어,영어,수학 평균 데이터
```

```python
df["학년"] = [3,3,2,1,1,3,2,2] #학년 Column 추가
df
```

```python
df.groupby(["학교","학년"]).mean() #학교별, 학년별 평균 데이터
```

```python
df.groupby("학년").mean()
```

```python
df.groupby("학년").mean().sort_values("키" ,ascending =False)
```

```python
df.groupby(["학교","학년"]).sum() #합계 
```

```python
df.groupby("학교")[["이름","SW특기"]].count() #학교로 그룹화를 한 뒤에 각 학교별 이름, SW특기 데이터의 수를 가져옴
```

```python
school = df.groupby("학교")
school["학년"].value_counts() #학교로 그룹화를 한 뒤 학년별 학생 수를 가져옴
```

```python
school["학년"].value_counts().loc["북산고등학교"] #학교로 그룹화를 한 뒤에 북산고에 대해서 학년별 학생수를 가져옴
```
