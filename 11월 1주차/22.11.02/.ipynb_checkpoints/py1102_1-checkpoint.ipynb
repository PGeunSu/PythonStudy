```python
#데이터 수정

import pandas as pd
df = pd.read_excel("score.xlsx", index_col="지원번호")
df
```

```python
df["학교"].replace({"북산고":'상봉고',"능남고":"한외고"})
```

```python
df["학교"].replace({"북산고":'상봉고',"능남고":"한외고"}, inplace=True)
df
```

```python
df["SW특기"] = df["SW특기"].str.lower()
df
```

```python
#Columns추가
df["총합"] = df["국어"] + df["영어"] + df["수학"] + df["과학"] + df["사회"]
df
```

```python
df["결과"] = "Fail" #결과 Column을 추가하고 전체데이터는 Fail로 초기화
df
```

```python
df.loc[df["총합"] > 400, "결과"] ="Pass"
df
```
