```python
import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure()
axes = figure.add_subplot()
x = [1,2,3,4]
y = [2,4,6,8]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,6])

axes.plot(x,y,color="red",linestyle="dashed",marker="^")
axes.plot(x2,y2,color="purple",linestyle="dotted",marker="o")

#linestyle : solid(기본값), dotted, dashdotted, dashed
#color : black, red ....
#marker : o , ^, > 등 값에 표시하는 마커의 형태
```

```python
import matplotlib.pyplot as plt
#x,y축을 기준으로 2개의 다른 축을 가질경우, 그래프 표현 가능
#twinx : x축을 기준으로 두개의 bar와 plot형태의 Y축 그래프 생성

figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x=[1,2,3,4]
y=[2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.bar(x,y,color="gray",label="bar")
axes2.plot(x2,y2,color="k",marker="o",label = "plot")
axes.legend()
axes2.legend(loc=1)
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes = figure.add_subplot(111)

x=[1,2,3,4]
y=[2,4,6,8]

x2 = [1,2,3,5]
y2 = [4,4,3,6]

axes.scatter(x,y)
axes.scatter(x2,y2)
```

```python
import matplotlib.pyplot as plt
#원그래프 - 데이터간의 비율을 파악하기 쉬움

figure = plt.figure()
axes = figure.add_subplot(111)

#label = ["a","b","c","d"]
label=['일','이','삼','사']
data = [1,1,2,3]

axes.pie(data,labels=label)
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
#autopct 부채꼴안에 표시될 숫자의 형식 지정

plt.pie(ratio, labels = labels, autopct="%.1f%%")
plt.show()
```
