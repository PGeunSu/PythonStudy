```python
import pandas as pd
data = pd.read_csv("C:\\Users\\KGA_17\\Downloads\\obs1.csv", encoding="cp949")
data = data[["지점명","일시","기온(°C)"]]
data
```

```python
#서울,부산 분리하기
is_seoul = data["지점명"] == "서울"
seoul = data[is_seoul]
seoul
```

```python
busan = data[data['지점명']=='부산']
busan
```

```python
temp = []
for i in busan['일시']:
    temp.append(i.split(' ')[0])
lack_busan = []
unique_temp = set(temp)
for i in unique_temp:
    if temp.count(i) != 24:
        lack_busan.append(i)
print(lack_busan)
```

```python
temp = []
for i in seoul["일시"]:
    temp.append(i.split(" ")[0])
lack_seoul = []
unique_temp = set(temp)
for i in unique_temp:
    if temp.count(i) != 24:
        lack_seoul.append(i)
lack_seoul
```

```python
lack_data = list(set(lack_seoul + lack_busan))
for i in lack_data:
    seoul = seoul[seoul["일시"].str.split(" ").str[0] != i]
    busan = busan[busan["일시"].str.split(" ").str[0] != i]
    print(len(seoul))
print(len(busan))
```
