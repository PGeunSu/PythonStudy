```python
import pandas as pd
file_name = "202108_202108_연령별인구현황_월간.xlsx"
df_m = pd.read_excel(file_name, skiprows=3, index_col ="행정기관", usecols="B,E:Y")
df_m
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns #컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(",","").astype(int) #1,195,1951 => 119051951 (정수형)
df_w
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(",","").astype(int) #1,195,1951 => 119051951 (정수형)
df_m
```

```python
#데이터 시각화
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"] = "AppleGothic"
matplotlib.rcParams["font.size"] = 15 # 글자크기
matplotlib.rcParams["axes.unicode_minus"] = False #한글 폰트 사용 시 마이너스 글저가 꺠지는 현상을 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) #단위 천명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title("2021대한민국 인구 피라미드")
plt.savefig("2021_인구피라미드.png", dpi=100)
plt.show()
```
